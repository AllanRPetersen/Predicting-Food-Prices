```python
NZ_food_prices_2020 = NZ_food_prices[NZ_food_prices['Period'] == 2020.01]
```

```python
NZ_food_prices_2020['Series_title_1'].unique()
```

```python
temp= NZ_food_prices_2020.copy()
```

```python
temp.tail()
```

```python
temp['Series_title_2'] = 0
```

```python
temp['Series_title_3'] = 0
```

```python
temp['Unit'] = 0
```

```python
for i, x in temp['Series_title_1'].iteritems():
    if x[-4:] == '.':
        temp['Unit'][i] = x[-5:]
    elif x[-4:] == 'each':
        temp['Unit'][i] = x[-4:]
    elif x[-4:] == 'pack':
        temp['Unit'][i] = x[-6:]
    elif x[-2:] == 'kg':
        temp['Unit'][i] = x[-3:]
    elif x[-2:] == 'ml':
        temp['Unit'][i] = x[-5:]
    elif x[-1:] == 'g':
        temp['Unit'][i] = x[-4:]
    elif x[-2:] == 're':
        temp['Unit'][i] = x[-7:]
    elif x[-3:] == 'res':
        temp['Unit'][i] = x[-8:]
```

```python
temp['Unit'].unique()
```

```python
for i, x in temp['Series_title_1'].iteritems():
    x_temp = x.split(', ')
    temp['Series_title_1'][i] = x_temp[0]
    temp['Series_title_2'][i] = x_temp[1]
    try:
        temp['Series_title_3'][i] = x_temp[2]
    except:
        continue
```

```python
temp.tail(30)
```

```python
temp.to_csv("NZ_food_prices_2020.csv")
```
